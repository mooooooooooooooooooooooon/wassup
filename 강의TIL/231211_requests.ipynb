### 상대 url을 절대 url로 변환하기

```python
href = sample['href']
```

```python
# //가 중복되는 문제처럼 생각보다 합치는게 번거로울 때가 있음

PATH + href
```

```python
from urllib.parse import urljoin
```

```python
urljoin(PATH, href)
```

```python
PATH +href[1:]
```

```python
news_title = []
news_url = []

for i in srclist :
    title = i.text
    url = urljoin(PATH, i['href'])
    print(title, url)
    news_title.append(title)
    news_url.append(url)
```

```python
news_title
```

```python
news_url
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({'제목' : news_title, '주소' : news_url})
df
```

```python
print(df)
```
